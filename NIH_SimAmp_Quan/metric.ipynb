```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load four audio files
filenames = [pd06_na, pd06_amp, pd06_70db, oc06_na, oc06_amp, oc06_70db]
titles = ["pd06_na", 'pd06_amp', 'pd06_70db', 'oc06_na', 'oc06_amp', 'oc06_70db']

# Create a new figure
plt.figure(figsize=(20, 10))

for i in range(6):
    # Load the audio file
    y, sr = librosa.load(filenames[i], sr=48000)

    # Compute the spectrogram
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)

    # Create a subplot and display the spectrogram
    plt.subplot(2, 3, i+1)
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='linear')
    plt.colorbar(format='%+2.0f dB')
    plt.title(titles[i])

plt.tight_layout()
plt.show()
```
